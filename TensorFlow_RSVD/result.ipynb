```python
import csv
import operator
import math
import random
from sets import Set
from numpy import genfromtxt, savetxt
from sklearn import cross_validation
from sys import argv
import numpy as np
import pandas as pd
import sys
import re

def read_process(filname, sep="\t"):
    col_names = ["user", "item", "rate", "st"]
    df = pd.read_csv(filname, sep=sep, header=None, names=col_names, engine='python')
    
    
    for col in ("user", "item"):
        df[col] = df[col].astype(np.int32)
    df["rate"] = df["rate"].astype(np.float32)
    return df

df = read_process("ratings.dat", sep="::")
print(df)
df.to_csv('ratings.csv',index=False,header=False)
f = open("ratings.csv")
ratings_data = csv.reader(f)
listData = list(ratings_data)
print(listData)
#f = open("ratings.csv")
#ratings_data = csv.reader(f)
#listData = list(ratings_data)


userRatings = {}

# User list without dictionary of movies and ratings 
users = sorted(listData, key=operator.itemgetter(0))

# mapping each user's movie preferences
count = 0
u_prev = 0;
for u in users:
    userID = u[0]
    movieID = u[1]
    movieRating = u[2]
    if(u_prev == userID):
        userRatings[userID][movieID] = movieRating
        u_prev = userID
    else:
        userRatings[userID] = {}
        userRatings[userID][movieID] = movieRating
        u_prev = userID




c = open("movies.csv")
movie = csv.reader(c)
listMovie = list(movie)
movieorder = sorted(listMovie, key=operator.itemgetter(0))
c.close()
#print(movieorder)
dic = {}
#transfer movie into {id : name,type}
for i in movieorder:
    dic[i[0]] = i[1]
#print(dic)
f2 = open("predictresult.csv")
total_data = csv.reader(f2)
totalData = list(total_data)
f2.close()
print(len(totalData))

for i in range(0,6040):
    uRatings = userRatings[str(i+1)]
    tmpdata = []
    for j in range(0,3883):
        if totalData[j+i*3883][1] in uRatings: continue
        tmpdata.append(totalData[j+i*3883])
    tmpdata = sorted(tmpdata, key=operator.itemgetter(2),reverse = True)
    print(len(tmpdata))
    print(tmpdata[0][0]+'::'+str(dic[tmpdata[0][1]])+'::'+str(dic[tmpdata[1][1]])+'::'+str(dic[tmpdata[2][1]])+'::'+str(dic[tmpdata[3][1]])+'::'+str(dic[tmpdata[4][1]])+'\n')
    file = open('recommend.txt', 'a')
    #file.write(tmpdata[0][0]+'::'+str(dic[tmpdata[0][1]])+'::'+str(dic[tmpdata[1][1]])+'::'+str(dic[tmpdata[2][1]])+'::'+str(dic[tmpdata[3][1]])+'::'+str(dic[tmpdata[4][1]])+'\n')
    #if tmpdata[0][0]==6040:
        #print(tmpdata[0][0]+'::'+str(dic[tmpdata[0][1]])+'::'+str(dic[tmpdata[1][1]])+'::'+str(dic[tmpdata[2][1]])+'::'+str(dic[tmpdata[3][1]])+'::'+str(dic[tmpdata[4][1]])+'\n')
print("finish")
```
